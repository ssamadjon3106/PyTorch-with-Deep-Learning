```python
tensor5=torch.rand(2, 3)
tensor5_to_gpu=tensor5.to(device)
print(tensor5_to_gpu)
torch.manual_seed(1234)
tensor6=torch.rand(2, 3)
tensor6_to_gpu=tensor6.to(device)
print(tensor6_to_gpu)
```

```python
# Perform matmul on tensor_A and tensor_B
output=torch.mm(tensor5, tensor6.T)
output, output.device, output.shape
```

```python
# Find max
max=torch.max(output)

# Find min
min=torch.min(output)
max, min
```

```python
argmax=torch.argmax(output)
argmin=torch.argmin(output)
argmax, argmin
```
